```python
#Program 51- House size

print("House floor area calculator.")
n=int(input("Please enter the number of rooms. "))
R=[]
A=[]
sum=0
for i in range(n):
    print("Please enter the name of room",i+1,":")
    r=input()
    print("Please enter the length(m) of room",i+1,":")
    l=float(input())
    print("Please enter the width(m) of room",i+1,":")
    w=float(input())
    R.append(r)
    a=l*w
    sum=sum+a
    A.append(round(a,1))
print("House details:")
for k in range(n):
    print("Room-",R[k])
    print(A[k],"metres squared")
print("The total area is calculated as:")
print(round(sum,1),"metres squared.")
```

```python
#PRogram 53- Bowling club day trip

m=int(input('How many members have requested a seat?'))
if m>20:
    print("Sorry, the maximum limit is 20.")
    print("How many members have requested a seat?")
    m=int(input())
a=[["empty"]*4 for main in range(5)]
for row in range(5):
    n=input('Please enter your name. ')
    r=int(input('Which row would you like to sit in? '))
    s=int(input('Which seat number would you like to sit in? '))
    if a[r-1][s-1]=='empty':
        a[r-1][s-1]=n
        print("Thank you, your name has been added.")
    else:
        print("Sorry, the seat is occupied.")
        r=int(input('Please enter row number'))
        s=int(input('Please enter seat number'))
        a[r-1][s-1]=n
print("Bus seats have been booked as follows:")
for i in range(5):
    for j in range(4):
        print(a[i][j],"",end='')
    print()
```

```python
#Program 54- Counting weeds

q=[[0]*5 for main in range(5)]
sum=0
for row in range(5):
    for column in range(5):
        q[row][column]=int(input('Please enter the number of weeds counted at '+str(row+1)+','+str(column+1)+': '))
        sum=sum+q[row][column]
print("The results are shown below:")
for row in range(5):
    for column in range(5):
        print(q[row][column],"",end='')
    print()
print("The total number of weeds found was -",str(sum)+".")
```
