```python
#Program 56- Guess the number part 2

import random
n=random.randint(1,100)
print('Guess the hidden number between 1 and 100.')

while True:
    num=int(input('Enter your guess.'))    
    if num>100 or num<1:
        print("Your guess was not valid. Enter it again")
        continue
    elif num>n:
        print("Your guess is too high. Try again.")
        continue
    elif num<n:   
        print('Your guess was too low. Try again')
        continue
    elif num==n:
        print('Correct, well done!')
        break
```

```python
#Program 58- Dress sizes

stock=[]
while True:
    q=input('Do you wish to add(A) or remove(R) a dress? ')
    
    if q=='A':
        print('Please enter a new dress size: ')
        sa=int(input())
        if sa==999:
            print("Stock=",stock)
            print("The largest dress in stock is size",max(stock),".")
            print("The smallest dress in stock in size",min(stock),".")
            break
        else:
            stock.append(sa)
            print("Stock=",stock)
    
    if q=='R':
        print('Please enter a new dress size: ')
        sr=int(input())
        stock.remove(sr)
        print("Stock=",stock)
        continue
    
```

```python
#Program 61- Diving scores

scores=[]
s=input("Please enter the judges' scores:")
r=float(input('Please enter the difficulty rating:'))
k=s.split(',')
for i in k:
    j=int(i)
    scores.append(j)
for m in range(2):
    ma=max(scores)
    mi=min(scores)
    scores.remove(ma)
    scores.remove(mi)
print("The three remaining judges scores are:")
for a in range(3):
    
    print(scores[a])
t=sum(scores)  

print("The diver's score is:",round(t*r,1))
```
